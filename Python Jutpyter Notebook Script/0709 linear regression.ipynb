```python
import numpy as np
from tensorflow.keras.optimizers import SGD # 경사하강법
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
x_train = np.array([1, 2, 3])
y_train = np.array([1, 2, 3])

model = Sequential() # 선형 회귀 모델
model.add(Dense(1, input_dim=1)) # input_dim 
# Dense = 퍼셉트론

sgd = SGD(lr=0.1) # lr : Learning Rate
model.compile(loss='mse', optimizer=sgd)
# loss = 비용함수, 비용함수로 Mean Sqaure Error 를 이용한다. 
# optimizer = 무엇으로 찾을거냐, sgd = 경사 하강법

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=200)
# epochs(에폭) 200번을 돌려요

y_predict = model.predict(np.array([5.5]))
print(y_predict)
```

```python
x_train = np.array([[73,80,75],[93,88,93],[89,91,90],[96,98,100],[73,66,70]])
y_train = np.array([152, 185, 180, 196, 142])

model = Sequential() # 선형 회귀 모델
model.add(Dense(2, input_dim=3)) # input_dim 
# Dense = 퍼셉트론

sgd = SGD(lr=0.00001) # lr : Learning Rate
model.compile(loss='mse', optimizer=sgd)
# loss = 비용함수, 비용함수로 Mean Sqaure Error 를 이용한다. 
# optimizer = 무엇으로 찾을거냐, sgd = 경사 하강법

# prints summary of the model to the terminal
model.summary()

model.fit(x_train, y_train, epochs=20)
# epochs(에폭) 200번을 돌려요
test = np.array([[70,80,90]]) ## 2차원으로 줬으면 test 데이터도 2차원으로 줘야됨
# np.expand_dims(test)
y_predict = model.predict(test)
print(y_predict)
```
