```python
df = pd.DataFrame({'name':['kim','lee','bae','cho','kang'],'sex':[1,2,2,1,1],'home':['마포','수지','처인','용산','종로'],
               'income':[250,195,325,210,274]})
num_to_str={1:'남',2:'여'}
to_city={'마포':'서울','용산':'서울','종로':'서울','처인':'용인','수지':'용인'}
```

```python
df
```

```python
def f(x):
    return x*0.015
df['gender']=df['sex'].map(num_to_str)
df['city']=df['home'].map(to_city)
df['new_income']=df['income'].map(f)
```

```python
df
```
