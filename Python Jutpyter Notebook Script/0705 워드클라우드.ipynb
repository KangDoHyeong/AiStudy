```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
STOPWORDS # 워드클라우드에 담지 않을 단어들
stopwords = set(STOPWORDS)
stopwords.add('state') #스탑워드에 state 추가
```

```python
alice_mask = np.array(Image.open('data/alice_mask.png'))
```

```python
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask,stopwords = stopwords).generate(text)
```

```python
r = lambda: np.random.randint(0,255)
def wc_color(word,font_size, position, orientation, randomstate=None, **kwargs):
    return (r(), r(), r())
plt.figure(figsize = (12,12))
plt.imshow(wc.recolor(color_func=wc_color, random_state=2), interpolation='bilinear')  #imageshow 이미지를 보여줍니다
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Okt
import matplotlib.pyplot as plt

from PIL import Image
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
```

```python
ok = Okt()
```

```python
ko_con_text= open('data/speech.txt',encoding='utf-8').read()
ko_con_text
```

```python
tokens_ko = ok.nouns(ko_con_text)
tokens_ko
```

```python
stop_words = list()
with open('data/stop_word.txt',encoding='utf-8') as f:
    for line in f.readlines():
        stop_words.append(line.strip())
stop_words
```

```python
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
tokens_ko # 삭제 할 단어가 들어있으면 tokens_ko에 돌려주지마라!
```

```python
sel_word = nltk.Text(tokens_ko)
data = sel_word.vocab().most_common(1000)
tmp_data = dict(data)
```
